```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im = cv.imread('D:\semester 5\EN3160\images\iman.jpg')
assert im is not None

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20,10))
ax.imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
ax.axis('off')
plt.show()
```

```python
import cv2 as cv
from matplotlib import cm
import numpy as np
import matplotlib.pyplot as plt

im = cv.imread('D:\semester 5\EN3160\images\iman.jpg' , cv.IMREAD_GRAYSCALE)
assert im is not None

im2 = im + 100

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
ax[0].imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB), cmap=cm.gray, vmin=0, vmax=255)
ax[0].axis('off')
ax[1].imshow(cv.cvtColor(im2, cv.COLOR_BGR2RGB),cmap=cm.gray, vmin=0, vmax=255)
ax[1].axis('off')
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im = cv.imread('D:\semester 5\EN3160\images\iman.jpg')
assert im is not None

# Resize image to 200x200 pixels
resized_image = cv.resize(im, (8000, 8000))


fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20,10))
ax.imshow(cv.cvtColor(resized_image, cv.COLOR_BGR2RGB))
ax.axis('off')
plt.show()
```
